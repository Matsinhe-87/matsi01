```python
import numpy as np
import matplotlib.pyplot as plt

vetor = np.array([4,8,12,20,30,30,0])
print(vetor)

# dobro do vetor
dobro_vetor = vetor*2
print(dobro_vetor)

# Criando o grafico
plt.plot(dobro_vetor, linestyle = "--")
plt.grid()
plt.show()

# Numpy com valores pre-definidos
# zeros.
matriz_zeros = np.zeros((2,2))
print(matriz_zeros)
```

```python
matrix_zeros = np.zeros((5,3,6))
display(matrix_zeros)
```

```python
# Uns
matrix_uns = np.ones((3,2,7))
print(matrix_uns)
```

```python
# Sequencias:
sequencia = np.arange(0,10,2)
print(sequencia)
```

```python
sequencias = np.arange(10,60,10)
print(sequencias)
```

```python
triplo_seq = [sequencias*3 for x in sequencias]
print(triplo_seq)
```

```python
seq = np.arange(15)
print(seq)
quad_seq = [seq**2 for i in seq]
print(quad_seq)
```

```python
# Espacamento igual:
espaco = np.linspace(0,100,5)
print(espaco)
```

```python
# Aleatorio
aleatorio = np.random.rand(3,2)
print(aleatorio)
```

```python
#Propriedades de arrays;
matriz = np.array([[2,4,6], [5,8,10], [11,4,9]])
print(matriz.shape)
print(matriz.shape,matriz.ndim,matriz.size)
print(matriz)
determinante =np.linalg.det(matriz)
print(determinante)
```
